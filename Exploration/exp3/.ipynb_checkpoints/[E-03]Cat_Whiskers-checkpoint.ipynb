```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
img = cv2.imread('./images/selfi.jpg')
img_show = img.copy()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
detector = dlib.get_frontal_face_detector()
dlib_rect = detector(img, 1)
```

```python
model_path = './models/shape_predictor_68_face_landmarks.dat'
predictor = dlib.shape_predictor(model_path)
```

```python
l = dlib_rect[0].left()
t = dlib_rect[0].top()
r = dlib_rect[0].right()
b = dlib_rect[0].bottom()
cv2.rectangle(img_show, (l, t), (r, b), (0, 255, 0), 2, lineType = cv2.LINE_AA)

points = predictor(img_rgb, dlib_rect[0])

list_points = list(map(lambda p : (p.x, p.y), points.parts()))

for point in list_points:
    cv2.circle(img_show, point, 2, (0, 255, 255), -1)
    
img_rgb_show = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb_show)
plt.show()
```

```python
print(list_points[29])
```

```python
x = list_points[30][0]
y = list_points[30][1]
h = list_points[33][1] - list_points[29][1]
w = h =  dlib_rect[0].width()  // 2
print('(x, y) : (%d, %d)'%(x, y))
print('(w, h) : (%d, %d)'%(w, h))
```

```python
sticker = cv2.imread('./images/cat.png')
print(sticker.shape)
sticker = cv2.resize(sticker, (w, h))
print(sticker.shape)
```

```python
refined_x = x - w // 2
refined_y = y - h // 2
```

```python
print(refined_x, refined_y)
```

```python
sticker_area = img_show[refined_y : refined_y + sticker.shape[0], refined_x : refined_x + sticker.shape[1]]
img_show[refined_y : refined_y + sticker.shape[0], refined_x : refined_x + sticker.shape[1]] = np.where(sticker == 255, sticker_area, sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_show_new = img.copy()
```

```python
sticker_area = img_show_new[refined_y : refined_y + sticker.shape[0], refined_x : refined_x + sticker.shape[1]]
img_show_new[refined_y : refined_y + sticker.shape[0], refined_x : refined_x + sticker.shape[1]] = np.where(sticker == 255, sticker_area, sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show_new, cv2.COLOR_BGR2RGB))
plt.show()
```

### 문제점

1. 얼굴이 돌아간 각도에 맞게 스티커의 각도도 변해야 한다. 그렇지 않으면 측면 얼굴에 정면 스티커가 붙어 적절하지 않게 된다.
2. 또한 현재 사용하는 라이브러리가 dlib.get_frontal_face_detector() 인데 얼굴이 조금만 측면으로 변하게 되면 제대로된 landmark detection의 정확도가 떨어지게 된다.
3. 멀리서 사용하는 경우 얼굴을 detection 하지 못한다.
4. 옆으로 누워서 촬영하게 되면 이또한 얼굴을 인식하지 못할 확률이 높을 뿐더러, 스티커의 모양도 변해야 하고 좌표 모든 값들이 변하게 된다.
5. 이러한 문제점은 face landmark detection 모델을 정면~측면에 이르는 모든 얼굴에 대해 새롭게 학습한 모델로 변경해서 진행해야 한다. 또한, 얼굴의 각도를 계산하여 스티커의 모양도 변경해주는 알고리즘이 필요할 것이다.(실제 서비스는 이를 AR 기술로 해결하고 있음)
6. 실제 서비스로 진행될 경우 실시간으로 사용자가 얼굴에 위치한 스티커를 보며 선택하고 촬영해야 하기 때문에, 실행속도는 매우 중요하다.
7. 스티커의 위치가 적절하게 있지 않게 되면 사용하지 않을 것이 분명하기 때문에, 정확도도 매우 중요하다.
